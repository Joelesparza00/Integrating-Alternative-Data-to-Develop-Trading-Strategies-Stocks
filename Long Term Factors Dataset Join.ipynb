```python
import pandas as pd
import numpy as np
```

```python
head = pd.read_csv('Key Developments Annual.csv')
executive = pd.read_csv('Executive Compensation Annual.csv')
annual = pd.read_csv('combined_annual_reports.csv')
```

```python
head['companyname'].unique()
```

```python
# Dictionary mapping company names to ticker symbols
company_to_ticker = {
    'Alphabet Inc.': 'googl',
    'Apple Inc.': 'aapl',
    'Eli Lilly and Company': 'lly',
    'J.P. Morgan Exchange-Traded Fund Trust - JPMorgan Market Expansion Enhanced Equity ETF': 'jpm',
    'J.P. Morgan Exchange-Traded Fund Trust - JPMorgan Social Advancement ETF': 'jpm',
    'J.P. Morgan Exchange-Traded Fund Trust - JPMorgan Sustainable Consumption ETF': 'jpm',
    'Meta Platforms, Inc.': 'meta',
    'Microsoft Corporation': 'msft',
    'NVIDIA Corporation': 'nvda',
    'Tesla, Inc.': 'tsla',
    'Walmart Inc.': 'wmt'
}

# Map company names to ticker symbols and create a new 'ticker' column
head['ticker'] = head['companyname'].map(company_to_ticker)

# Display the DataFrame with the new 'ticker' column
print(head.head())
```

```python
executive.head()
```

```python
# Assuming "executive" is your DataFrame containing the dataset
executive['TICKER'] = executive['TICKER'].str.lower()

# Display the updated DataFrame
print(executive.head())
```

```python
annual.head()
```

```python

# Merge "head" and "annual" datasets on "ticker" and "year"
merged_head_annual = pd.merge(head, annual, left_on=['ticker', 'year'], right_on=['ticker', 'year'], how='inner')

# Merge "executive" and "annual" datasets on "TICKER" and "YEAR"
merged_executive_annual = pd.merge(executive, annual, left_on=['TICKER', 'YEAR'], right_on=['ticker', 'year'], how='inner')

# Merge "head", "executive", and "annual" datasets
merged_all = pd.merge(merged_head_annual, merged_executive_annual, left_index=True, right_index=True, how='inner')

# Display the first few rows of the merged dataset
print(merged_all.head())
```
