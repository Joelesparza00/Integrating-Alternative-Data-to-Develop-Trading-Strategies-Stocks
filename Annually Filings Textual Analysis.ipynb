## Apple 2023 Annual Report Textual Analysis

```python
import requests
import pdfplumber
import tempfile
import os
import re

# URL of the PDF file
url = "https://d18rn0p25nwr6d.cloudfront.net/CIK-0000320193/faab4555-c69b-438a-aaf7-e09305f87ca3.pdf"

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Create a temporary file to save the PDF content
    with tempfile.NamedTemporaryFile(delete=False) as tf:
        tf.write(response.content)
        pdf_path = tf.name

    # Open the PDF file
    with pdfplumber.open(pdf_path) as pdf:
        # Extract text from each page
        text = ""
        for page in pdf.pages[4:]:
            page_text = page.extract_text()
            # Extract only paragraphs (lines with more than 20 characters)
            paragraphs = re.findall(r'.{20,}', page_text)
            text += '\n'.join(paragraphs) + '\n'
    
    # Maximum number of tokens for the prompt
    max_prompt_tokens = 10000  

    # Trim the text to fit within the maximum length
    trimmed_text2 = text[:max_prompt_tokens]

    # Print results
    print(trimmed_text2)
    
    # Remove the temporary PDF file
    os.unlink(pdf_path)
else:
    print("Failed to download the PDF file")
```

```python
import requests
import pdfplumber
import tempfile
import os
import re
import pandas as pd
from openai import OpenAI

# Define an empty list to store the results
results = []

# Create an instance of the OpenAI client
api_key = "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"
client = OpenAI(api_key=api_key)

# Iterate over each URL
for year, url in enumerate(reversed(aapl), start=2013):
    # Send a GET request to the URL
    response = requests.get(url)
    
    # Check if the request was successful
    if response.status_code == 200:
        # Create a temporary file to save the PDF content
        with tempfile.NamedTemporaryFile(delete=False) as tf:
            tf.write(response.content)
            pdf_path = tf.name

        # Open the PDF file
        with pdfplumber.open(pdf_path) as pdf:
            # Extract text from each page
            text = ""
            for page in pdf.pages[4:]:
                page_text = page.extract_text()
                # Extract only paragraphs (lines with more than 20 characters)
                paragraphs = re.findall(r'.{20,}', page_text)
                text += '\n'.join(paragraphs) + '\n'
        
        # Maximum number of tokens for the prompt
        max_prompt_tokens = 7500  

        # Trim the text to fit within the maximum length
        trimmed_text = text[:max_prompt_tokens]
        
        # Create a request to the Completions endpoint
        completion = client.completions.create(
            model="gpt-3.5-turbo-instruct",
            prompt=f"Count the sentences that explicitly reference the company's culture. Provide a numeric value as the output. For example, 18: {trimmed_text}",
            max_tokens=1000
        )
        
        # Extract the count from the response using regular expressions
        pattern = r'(\d+)'
        matches = re.findall(pattern, completion.choices[0].text.strip())
        count = int(matches[0]) if matches else None
        
        # Append the results to the list
        results.append({'ticker': 'AAPL', 'year': year, 'count': count})
        
        # Remove the temporary PDF file
        os.unlink(pdf_path)
    else:
        print(f"Failed to download the PDF file for year {year}")

# Create a dataframe from the results list
apple = pd.DataFrame(results)

# Print the dataframe
print(apple)
```
