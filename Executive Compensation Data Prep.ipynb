```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import statsmodels.api as sm
from scipy import stats
```

```python
executive = pd.read_csv('executive compensation.csv')
print(executive.head(5))
```

```python
# Initialize a new DataFrame to store the calculated tenure
tenure_df = pd.DataFrame(columns=['EXEC_FULLNAME', 'TICKER', 'YEAR', 'TENURE'])

# Get unique tickers
unique_tickers = executive['TICKER'].unique()

# Iterate over each unique ticker
for ticker in unique_tickers:
    ticker_data = executive[executive['TICKER'] == ticker]
    exec_tenure = {}  # Dictionary to store tenure for each executive for the current ticker
    
    # Iterate over each row for the current ticker
    for _, row in ticker_data.iterrows():
        exec_name = row['EXEC_FULLNAME']
        year = row['YEAR']
        
        # If the executive is not in the exec_tenure dictionary for the current ticker, add them with tenure 1
        if exec_name not in exec_tenure:
            exec_tenure[exec_name] = 1
        else:
            # Check if the executive appeared in the previous year
            prev_year = year - 1
            if prev_year in ticker_data[ticker_data['EXEC_FULLNAME'] == exec_name]['YEAR'].values:
                exec_tenure[exec_name] += 1
            else:
                # If the executive did not appear in the previous year, reset their tenure to 1
                exec_tenure[exec_name] = 1
        
        # Append the row with the calculated tenure to the tenure DataFrame
        tenure_df = tenure_df.append({'EXEC_FULLNAME': exec_name, 'TICKER': ticker, 'YEAR': year, 'TENURE': exec_tenure[exec_name]}, ignore_index=True)

# Merge the calculated tenure with the original dataframe on 'EXEC_FULLNAME' and 'TICKER'
executive = executive.merge(tenure_df, on=['EXEC_FULLNAME', 'TICKER', 'YEAR'], how='left')

print(executive)
```

```python
executive_comp = executive[executive["YEAR"] >= 2014]
executive_comp.head(15)
```

```python
import numpy as np

# Clean 'TENURE' column and convert to numeric
executive_comp['TENURE'] = pd.to_numeric(executive_comp['TENURE'], errors='coerce')

# Clean 'Income Stability Ratio' column and convert to numeric
executive_comp['Income Stability Ratio'] = pd.to_numeric(executive_comp['Income Stability Ratio'].str.replace(r'[^0-9.]', ''), errors='coerce')

# Define the aggregation functions for each column
aggregations = {
    'CONAME': 'first',
    'SALARY': 'mean',
    'BONUS': 'mean',
    'TDC1': 'mean',
    'SAL_PCT': 'mean',
    'TDC1_PCT': 'mean',
    'GVKEY': 'first',
    'YEAR': 'first',
    'PAGE': 'mean',
    'Income Stability Ratio': 'mean',
    'TENURE': 'mean'
}

# Filter out non-numeric columns for aggregation functions that require numeric values
numeric_columns = ['SALARY', 'BONUS', 'TDC1', 'SAL_PCT', 'TDC1_PCT', 'PAGE', 'Income Stability Ratio', 'TENURE']
aggregations_numeric = {key: value for key, value in aggregations.items() if key in numeric_columns}

# Group by 'TICKER' and 'YEAR', then apply the aggregations
grouped_executive = executive_comp.groupby(['TICKER', 'YEAR']).agg(aggregations_numeric)

# Flatten the column names
grouped_executive.columns = ['_'.join(col).strip() for col in grouped_executive.columns.values]

# Reset index to make the groupby columns regular columns
grouped_executive.reset_index(inplace=True)

print(grouped_executive)
```
