```python
#!/usr/bin/env python
from urllib.request import urlopen
import certifi
import json

def get_jsonparsed_data(url):
    response = urlopen(url, cafile=certifi.where())
    data = response.read().decode("utf-8")
    return json.loads(data)

# url = (f"https://financialmodelingprep.com/api/v3/stock_news?page=0&apikey={FMP_API_KEY}")
url = (f"https://financialmodelingprep.com/api/v3/fmp/articles?page=0&size=5&apikey={FMP_API_KEY}")

d = get_jsonparsed_data(url)
```

```python
# Calculate positive and negative sentiment score using OpenAI API
from openai import OpenAI
client = OpenAI(api_key="")

text = d["content"][0]["content"]

# Create a request to the Completions endpoint
response = client.completions.create(
  model="gpt-3.5-turbo-instruct",
  prompt=f"classify the sentiment of the following statements as either negative, positive, or neutral: {text}",
  max_tokens=100
)
print(response.choices[0].text)
```

```python
import time
import pandas as pd

def get_sentiment_for_article(article_text):
  response = client.completions.create(
      model="gpt-3.5-turbo-instruct",
      prompt=f"Compulsorily classify the following statement (entire) as either one of these - negative, positive, or neutral: {article_text}",
      max_tokens=100)

  sentiment = response.choices[0].text
  print(f"Sentiment: {sentiment}, Article: {article_text}")
  return sentiment.split("\n")[-1]

all = list()

for i in range(50):

  url = (f"https://financialmodelingprep.com/api/v3/fmp/articles?page={str(i)}&size=500&apikey={FMP_API_KEY}")
  d = get_jsonparsed_data(url)

  for d_i in d["content"]:
    d_i["sentiment"] = get_sentiment_for_article(d_i["content"])

  d = pd.DataFrame(d["content"])
  all.append(d)

  print(f"Iteration {i} complete")

  time.sleep(2)

df = pd.concat(all, ignore_index=True).reset_index(drop=True)
print(df.head())
df.to_csv("trail_data.csv")
```
